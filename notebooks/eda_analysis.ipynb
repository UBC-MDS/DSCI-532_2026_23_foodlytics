## NUMBER OF RESTAURANTS IN EACH CANADIAN CITY

```python
alt.Chart(df).mark_bar().encode(x= alt.X("city", sort="-y", title="City", axis=alt.Axis(labelAngle=0)),
                                y= alt.Y("count()", title="Number of Restaurants"),
                                color= alt.Color("city", 
                                                 legend=alt.Legend(title="Canadian city")),
                                tooltip =["city"]).properties(
        title="Number of restaurants in the center of each city",
        width=500, height=400
    
    )
```

Toronto has the highest number of restaurants in the city center, followed by Vancouver and Montreal. Calgary and Edmonton are in the middle range, while Ottawa, Winnipeg, and especially Brampton have much fewer restaurants.

This suggests that bigger cities have a more saturated and competitive market, while smaller cities may present less competition but possibly lower demand.

## NUMBER OF REVIEWS OF EACH CITY

```python
alt.Chart(df).mark_bar().encode(
    x=alt.X("city", sort="-y", axis=alt.Axis(labelAngle=0)),
    y=alt.Y("sum(num_reviews)", title="Total Number of Reviews")
).properties(
    title="Number of reviews for each city", width=500
)
```

Vancouver shows the highest total number of reviews, followed by Toronto and Calgary. Montreal and Winnipeg are in the middle range, while Edmonton, Ottawa, and especially Brampton have significantly fewer reviews.

This suggests that customer engagement is much higher in Vancouver and Toronto, which may indicate stronger demand but also higher competition. Smaller cities like Brampton and Ottawa appear to have lower overall activity in terms of reviews.

## AVERAGE STAR IN EACH CITY

```python
chart =alt.Chart(df).mark_bar().encode(
    y=alt.Y("city:N", sort="x", title=None),
    x=alt.X(
        "mean(star):Q",
        scale = alt.Scale(domain=[3.5, 5]),
        title="Average Star Rating"
    ),
    color="city:N",
    tooltip=["city", "mean(star)"]
).properties(
    width=700,
    height=300,
    title="Average of stars in each city"
)

text= chart.mark_text(
    align="left",
    dx=5
).encode(
    text=alt.Text("mean(star):Q", format=".2f")
)

chart + text
```

## MOST COMMON FOOD CATEGORIES IN EACH CITY

```python

top5 = (
    df.groupby("city")["restaurant_category"]
      .value_counts()
      .groupby(level=0)
      .head(5)
      .reset_index(name="count")
)

base = alt.Chart(top5).mark_bar().encode(
    y=alt.Y(
        "restaurant_category:N",
        sort=alt.SortField("count", order="descending"),
        title=None
    ),
    x=alt.X("count:Q", title="Number of Restaurants"),
    tooltip=["city:N", "restaurant_category:N", "count:Q"]
).properties(
    width=320,
    height=220
)

base.facet(
    facet=alt.Facet("city:N", title=None),
    columns=2
).resolve_scale(
    y='independent'
)

```

Overall, Asian cuisines (especially Chinese and Japanese) are very strong across most major cities, particularly in Toronto and Vancouver. 

Vancouver clearly stands out for Japanese food, while Montreal shows a strong Italian presence. Brampton has a noticeable concentration of Indian restaurants, suggesting a strong local demand. 

Cities like Ottawa and Calgary appear more balanced, without one single cuisine dominating the market. Winnipeg shows more casual and comfort-oriented categories such as Breakfast and Bakery. 

The analysis of the most common restaurant categories shows which types of cuisine are most popular in each city. A high number of restaurants in one category suggests strong local demand, but it can also mean high competition. Therefore, cuisines that are popular overall but less common in certain cities may represent good opportunities for opening a new restaurant.

# EVALUATION OF THE PRICE RANGE 

```python
price_order = ["$", "$$", "$$$", "$$$$"]

df["price_range"] = pd.Categorical(df["price_range"], categories=price_order, ordered=True)
 
city_ranges = df.groupby("city")["price_range"].value_counts().sort_index()
city_ranges
```

From this table we can clearly see that in all cities the majority of restaurants are in the price range. This is by far the most common category everywhere. The $ category usually comes next, but it is still much smaller compared to.

The higher price ranges (and $$$$) are very rare in almost all cities. Toronto is the only city with 2 restaurants in the $$$$ category, followed by Brampton with only 1. The other cities have either zero or very few and no $$$$ at all.

So overall, the market is clearly dominated by mid-range ($$) restaurants, while premium options ($$$ and $$$$) are very limited.
